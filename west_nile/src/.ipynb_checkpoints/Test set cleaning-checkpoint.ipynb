## Test set cleaning

```python
# Load test set
test_set = pd.read_csv('../input/test.csv',na_values=['M'])
test_set.head()
```

```python
# Drop irrelevant columns
test_set = test_set.drop(['Address','Block','Street','Trap','AddressNumberAndStreet','AddressAccuracy'], axis=1)
```

```python
# Create month column
test_set['Date'] = pd.to_datetime(test_set['Date'])
test_set['month'] = test_set['Date'].map(lambda x: x.month)
test_set.head()
```

```python
# Create species dummies
test_set = pd.get_dummies(test_set, columns=['Species'])
test_set= test_set.drop('Species_UNSPECIFIED CULEX', axis=1)
test_set.head()
```
